```python
import numpy as np

def shuffle_and_split_data(X, Y, test_ratio, val_ratio):
    # Shuffle indices
    shuffled_indices = np.random.permutation(len(X))
    
    # Calculate the sizes for test and validation sets
    test_set_size = int(len(X) * test_ratio)
    val_set_size = int(len(X) * val_ratio)
    
    # Get indices for each set
    test_indices = shuffled_indices[:test_set_size]
    val_indices = shuffled_indices[test_set_size:test_set_size + val_set_size]
    train_indices = shuffled_indices[test_set_size + val_set_size:]
    
    # Split the data
    return (X[train_indices], X[val_indices], X[test_indices], 
            Y[train_indices], Y[val_indices], Y[test_indices])
```

```python
Xtrain_set, Xval_set, Xtest_set, Ytrain_set, Yval_set, Ytest_set = shuffle_and_split_data(Xarr, Yarr, test_ratio=0.15, val_ratio=0.15)
```

```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
```

```python
import xgboost as xgb

# Prepare data for XGBoost
xgb_train = xgb.DMatrix(Xtrain_set, label=Ytrain_set, enable_categorical=True)
xgb_val = xgb.DMatrix(Xval_set, label=Yval_set, enable_categorical=True)
xgb_test = xgb.DMatrix(Xtest_set, label=Ytest_set , enable_categorical=True)

# Define XGBoost parameters for multi-class classification
num_classes = len(set(Ytrain_set))  # Automatically detect the number of classes

params = {
    'objective': 'multi:softmax',  # Use 'multi:softmax' for direct class prediction
    'eval_metric': 'mlogloss',     # Multi-class log loss
    'max_depth': 5,
    'learning_rate': 0.01,
    'n_estimators': 100,
    'num_class': num_classes        # Specify the number of classes
}

# Training with early stopping
bst = xgb.train(
    params,
    xgb_train,
    num_boost_round=500,            # Number of boosting rounds
    early_stopping_rounds=10,       # Early stopping using validation set
    evals=[(xgb_val, "validation")]
)

y_val_pred = bst.predict(xgb_val)
y_test_pred = bst.predict(xgb_test)
```

```python
def accuracy(y_true, y_pred):
    TP, FP, TN, FN = confusion_matrix(y_true, y_pred)
    return (TP + TN) / (TP + TN + FP + FN)
```

```python
def confusion_matrix(y_true, y_pred):
    TP = sum((y_true == 1) & (y_pred == 1))
    FP = sum((y_true == 0) & (y_pred == 1))
    TN = sum((y_true == 0) & (y_pred == 0))
    FN = sum((y_true == 1) & (y_pred == 0))
    return TP, FP, TN, FN
```

```python
def precision(y_true, y_pred):
    TP, FP, TN, FN = confusion_matrix(y_true, y_pred)
    return TP / (TP + FP) if (TP + FP) != 0 else 0
```

```python
def recall(y_true, y_pred):
    TP, FP, TN, FN = confusion_matrix(y_true, y_pred)
    return TP / (TP + FN) if (TP + FN) != 0 else 0
```

```python
def f_score(y_true, y_pred):
    prec = precision(y_true, y_pred)
    rec = recall(y_true, y_pred)
    return 2 * (prec * rec) / (prec + rec) if (prec + rec) != 0 else 0
```

```python
def average_precision(y_true, y_pred, class_labels):
    precision_vals = []
    for label in class_labels:
        # Create binary classification for the current class
        binary_true = (y_true == label).astype(int)
        binary_pred = (y_pred == label).astype(int)
        
        # Calculate precision and recall for this class
        prec = precision(binary_true, binary_pred)
        recall_val = recall(binary_true, binary_pred)
        
        # Store precision for the current class
        precision_vals.append(prec)
    
    return sum(precision_vals) / len(precision_vals)
```

```python
def average_recall(y_true, y_pred, class_labels):
    recall_vals = []
    for label in class_labels:
        # Create binary classification for the current class
        binary_true = (y_true == label).astype(int)
        binary_pred = (y_pred == label).astype(int)
        
        # Calculate recall for this class
        recall_val = recall(binary_true, binary_pred)
        
        # Store recall for the current class
        recall_vals.append(recall_val)
    
    return sum(recall_vals) / len(recall_vals)
```
