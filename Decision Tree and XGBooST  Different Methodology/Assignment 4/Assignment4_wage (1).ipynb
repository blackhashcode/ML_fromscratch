```python
from sklearn.preprocessing import LabelEncoder

# Copy the original DataFrame
df_new = df.copy()

# Create a LabelEncoder object
label_encoder = LabelEncoder()

# Apply Label Encoding to each specified column
for column in ['maritl','race','education','region','jobclass','health','health_ins','logwage','wage']:
    df_new[column] = label_encoder.fit_transform(df_new[column])

df_new
```

```python
df_new.columns
```

```python
df_new = (df_new-df_new.min())/(df_new.max()-df_new.min())
df_new
```

```python
df_new['region'].isnull()
```

```python
df_new = df_new.drop(columns=['region'])
df_new
```

```python
y = df_new['wage']
X = df_new.drop(columns=['wage', 'logwage'])

Yarr = y.to_numpy()
Xarr = X.to_numpy()

print("Shape of Y:", Yarr.shape)
print("Shape of X:", Xarr.shape)
```

```python
import numpy as np

def shuffle_and_split_data(X, Y, test_ratio, val_ratio):
    # Shuffle indices
    shuffled_indices = np.random.permutation(len(X))
    
    # Calculate the sizes for test and validation sets
    test_set_size = int(len(X) * test_ratio)
    val_set_size = int(len(X) * val_ratio)
    
    # Get indices for each set
    test_indices = shuffled_indices[:test_set_size]
    val_indices = shuffled_indices[test_set_size:test_set_size + val_set_size]
    train_indices = shuffled_indices[test_set_size + val_set_size:]
    
    # Split the data
    return (X[train_indices], X[val_indices], X[test_indices], 
            Y[train_indices], Y[val_indices], Y[test_indices])
```

```python
Xtrain_set, Xval_set, Xtest_set, Ytrain_set, Yval_set, Ytest_set = shuffle_and_split_data(Xarr, Yarr, test_ratio=0.15, val_ratio=0.15)
```

```python
print("Shape of Xtrain_set:", Xtrain_set.shape)
print("Shape of Ytrain_set:", Ytrain_set.shape)
```

```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeRegressor ,plot_tree
import matplotlib.pyplot as plt
```

```python
clf = DecisionTreeRegressor(max_depth=5,random_state=42)
```

```python
clf.fit(Xtrain_set,Ytrain_set)
```

```python
y_val_pred = clf.predict(Xval_set)
y_test_pred = clf.predict(Xtest_set)
```

```python
def calculate_mse(y_true, y_pred):
    n = len(y_true)
    mse = sum((y_t - y_p) ** 2 for y_t, y_p in zip(y_true, y_pred)) / n
    return mse
    
mse1 = calculate_mse(Yval_set, y_val_pred)
print("Mean Squared Error:", mse1)
```

```python
mse2 = calculate_mse(Ytest_set, y_test_pred)
print("Mean Squared Error:", mse2)
```

```python
from xgboost import XGBRegressor
model = XGBRegressor()

model.fit(Xtrain_set, Ytrain_set)
y_train_pred_xgb = model.predict(Xtrain_set)
y_val_pred_xgb = model.predict(Xval_set)
y_test_xgb = model.predict(Xtest_set)
```

```python
mse3 = calculate_mse(Ytrain_set, y_train_pred_xgb)
print("Mean Squared Error:", mse3)
```
