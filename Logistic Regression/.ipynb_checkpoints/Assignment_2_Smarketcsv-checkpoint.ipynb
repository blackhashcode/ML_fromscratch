```python
import numpy as np
import pandas as pd
import matplotlib.pyplot
```

```python
df = pd.read_csv("Smarket.csv")
df
```

```python
df.isnull()
```

```python
df["Direction_Binary"] = df["Direction"].apply(lambda x: 1 if x == "Up" else 0)
```

```python
df = df.drop(columns=["Direction"])
```

```python
df_new = df.drop("Year", axis=1)
df_new
```

```python
df_normal = (df_new-df_new.min())/(df_new.max()-df_new.min())
df_normal
```

```python
Y = df_normal[["Direction_Binary"]]
X = df_normal.drop(columns=["Direction_Binary"])

Yarr = Y.to_numpy()
Xarr = X.to_numpy()

print("Shape of Y:", Yarr.shape)
print("Shape of X:", Xarr.shape)
```

```python
def shuffle_and_split_data(X, Y, test_ratio):
    shuffled_indices = np.random.permutation(len(X))
    test_set_size = int(len(X) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    
    return X[train_indices], X[test_indices], Y[train_indices], Y[test_indices]

Xtrain_set, Xtest_set, Ytrain_set, Ytest_set = shuffle_and_split_data(Xarr, Yarr, 0.25)
```

```python
print(Xtrain_set.shape)
print(Ytrain_set.shape)
print(Xtest_set.shape)
print(Ytest_set.shape)
```

```python
# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Cost function (Binary Cross-Entropy)
def compute_cost(X, Y, weights):
    m = len(Y)
    z = np.dot(X, weights)
    predictions = sigmoid(z)
    cost = -(1 / m) * np.sum(Y * np.log(predictions) + (1 - Y) * np.log(1 - predictions))
    return cost

# Gradient Descent
def gradient_descent(X, Y, weights, learning_rate, iterations):
    m = len(Y)
    for i in range(iterations):
        # Calculate predictions
        z = np.dot(X, weights)
        predictions = sigmoid(z)

        # Compute gradient
        gradient = (1 / m) * np.dot(X.T, (predictions - Y))

        # Update weights
        weights -= learning_rate * gradient

        # Print cost every 100 iterations
        if i % 100 == 0:
            cost = compute_cost(X, Y, weights)
            print(f"Iteration {i}, Cost: {cost}")
    return weights

```

```python
def predict(X, weights):
    probabilities = sigmoid(np.dot(X, weights))
    return [1 if prob >= 0.5 else 0 for prob in probabilities]
```

```python
print(Xtrain_set.shape[1])
```

```python
Ytrain_set = Ytrain_set.ravel()  
Ytest_set = Ytest_set.ravel()    
```

```python
# Initialize weights
weights = np.zeros(Xtrain_set.shape[1])

weights = gradient_descent(Xtrain_set, Ytrain_set, weights, 0.2 , 1000)
```

```python
# Test the model
predictions = predict(Xtrain_set, weights)

# Evaluate the model
accuracy = np.mean(predictions == Ytrain_set) * 100
print(f"Training Model Accuracy: {accuracy:.2f}%")
```

```python
# Test the model
predictions = predict(Xtest_set, weights)

# Evaluate the model
accuracy = np.mean(predictions == Ytest_set) * 100
print(f"Testing Model Accuracy: {accuracy:.2f}%")
```
