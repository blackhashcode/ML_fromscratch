```python
df = pd.read_csv("AirQualityUCI.csv", delimiter=';')
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
dataset= df.dropna() 
dataset.replace(-200, np.nan, inplace=True)
dataset.dropna(inplace=True)
dataset
```

```python
import datetime

dataset = pd.DataFrame(dataset)
dataset['Date'] =dataset['Date'].apply(pd.to_datetime, format='mixed')
dataset['Time'] =dataset['Time'].astype('category').cat.codes
dataset['CO(GT)'] = dataset['CO(GT)'].str.replace(',', '.').astype(float)
dataset['PT08.S1(CO)'] =dataset['PT08.S1(CO)'].astype('category').cat.codes
dataset['NMHC(GT)'] =dataset['NMHC(GT)'].astype(int)
dataset['C6H6(GT)'] =dataset['C6H6(GT)'].str.replace(',', '.').astype(float)
dataset['PT08.S2(NMHC)'] =dataset['PT08.S2(NMHC)'].astype('category').cat.codes
dataset['NOx(GT)'] =dataset['NOx(GT)'].astype(int)
dataset['PT08.S3(NOx)'] =dataset['PT08.S3(NOx)'].astype('category').cat.codes
dataset['NO2(GT)'] =dataset['NO2(GT)'].astype(int)
dataset['PT08.S4(NO2)'] =dataset['PT08.S4(NO2)'].astype('category').cat.codes
dataset['PT08.S5(O3)'] =dataset['PT08.S5(O3)'].astype('category').cat.codes
dataset['T'] =dataset['T'].str.replace(',', '.').astype(float)
dataset['RH'] =dataset['RH'].str.replace(',', '.').astype(float)
dataset['AH'] =dataset['AH'].str.replace(',', '.').astype(float)
```

```python
Y = data_normal['CO(GT)']
X = data_normal.drop(columns=['CO(GT)'])

Yarr = Y.to_numpy()
Xarr = X.to_numpy()

print("Shape of Y:", Yarr.shape)
print("Shape of X:", Xarr.shape)
```

```python
def shuffle_and_split_data(X, Y, test_ratio):
    shuffled_indices = np.random.permutation(len(X))
    test_set_size = int(len(X) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    
    return X[train_indices], X[test_indices], Y[train_indices], Y[test_indices]

Xtrain_set, Xtest_set, Ytrain_set, Ytest_set = shuffle_and_split_data(Xarr, Yarr, 0.25)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Initialize parameters
def initialize_parameters(lenw):
    w = np.random.randn(1, lenw)
    b = 0
    return w, b

# Forward propagation
def forward_prop(X, w, b):
    z = np.dot(w, X) + b  # z has shape (1, m)
    return z

# Cost function (Mean Squared Error)
def cost_function(z, y):
    m = y.shape[1]
    J = (1 / (2 * m)) * np.sum(np.square(z - y))
    return J

# Backward propagation
def back_prop(X, y, z):
    m = y.shape[1]
    dz = (1 / m) * (z - y)
    dw = np.dot(dz, X.T)
    db = np.sum(dz)
    return dw, db

# Gradient descent update
def gradient_descent_update(w, b, dw, db, learning_rate):
    w = w - learning_rate * dw
    b = b - learning_rate * db
    return w, b

# Linear regression model
def linear_regression_model(Xtrain, ytrain, Xtest, ytest, learning_rate, epochs):
    lenw = Xtrain.shape[0]  # Number of features
    w, b = initialize_parameters(lenw)

    costs_train = []  # Store training costs
    m_train = ytrain.shape[0]  # Number of training samples
    m_test = ytest.shape[0]  # Number of testing samples

    # Reshape ytrain and ytest to match the shape of the predictions (1, m)
    ytrain = ytrain.reshape(1, -1)  # Reshape to (1, m_train)
    ytest = ytest.reshape(1, -1)  # Reshape to (1, m_test)

    for i in range(1, epochs + 1):
        # Forward propagation for training set
        z_train = forward_prop(Xtrain, w, b)
        cost_train = cost_function(z_train, ytrain)

        # Backward propagation
        dw, db = back_prop(Xtrain, ytrain, z_train)

        # Gradient descent update
        w, b = gradient_descent_update(w, b, dw, db, learning_rate)

        if i % 10 == 0:
            costs_train.append(cost_train)

        # **Mean Squared Error (MSE) for training and testing**
        MSE_train = (1 / m_train) * np.sum(np.square(z_train - ytrain))

        # Testing on the test set
        z_test = forward_prop(Xtest, w, b)
        cost_test = cost_function(z_test, ytest)
        MSE_test = (1 / m_test) * np.sum(np.square(z_test - ytest))

        # Print progress every 10 epochs
        if i % 10 == 0:
            print(f'Epoch {i}/{epochs}:')
            print(f'Training cost: {cost_train} | Test cost: {cost_test}')
            print(f'MSE train: {MSE_train} | MSE test: {MSE_test}')

    # Plot the training cost over time
    plt.plot(costs_train)
    plt.xlabel('Iterations (per tens)')
    plt.ylabel('Training Cost')
    plt.title(f'Learning rate: {learning_rate}')
    plt.show()
```

```python
linear_regression_model(Xtrain_set.T, Ytrain_set, Xtest_set.T, Ytest_set, 0.5, 1000)
```
