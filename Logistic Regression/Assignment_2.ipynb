```python
def shuffle_and_split_data(X, Y, test_ratio):
    shuffled_indices = np.random.permutation(len(X))
    test_set_size = int(len(X) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    
    return X[train_indices], X[test_indices], Y[train_indices], Y[test_indices]

Xtrain_set, Xtest_set, Ytrain_set, Ytest_set = shuffle_and_split_data(Xarr, Yarr, 0.25)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Initialize parameters
def initialize_parameters(lenw):
    w = np.random.randn(1, lenw)
    b = 0
    return w, b

# Forward propagation
def forward_prop(X, w, b):
    z = np.dot(w, X) + b  # z has shape (1, m)
    return z

# Cost function with L2 Regularization (Mean Squared Error + Regularization term)
def cost_function(z, y, w, lambd):
    m = y.shape[1]
    J = (1 / (2 * m)) * np.sum(np.square(z - y)) + (lambd / (2 * m)) * np.sum(np.square(w))
    return J

# Backward propagation with L2 regularization
def back_prop(X, y, z, w, lambd):
    m = y.shape[1]
    dz = (1 / m) * (z - y)
    dw = np.dot(dz, X.T) + (lambd / m) * w  # Add regularization term to dw
    db = np.sum(dz)
    return dw, db

# Gradient descent update
def gradient_descent_update(w, b, dw, db, learning_rate):
    w = w - learning_rate * dw
    b = b - learning_rate * db
    return w, b

# Linear regression model with regularization
def linear_regression_model(Xtrain, ytrain, Xtest, ytest, learning_rate, epochs, lambd):
    lenw = Xtrain.shape[0]  # Number of features
    w, b = initialize_parameters(lenw)

    costs_train = []  # Store training costs
    m_train = ytrain.shape[0]  # Number of training samples
    m_test = ytest.shape[0]  # Number of testing samples

    # Reshape ytrain and ytest to match the shape of the predictions (1, m)
    ytrain = ytrain.reshape(1, -1)  # Reshape to (1, m_train)
    ytest = ytest.reshape(1, -1)  # Reshape to (1, m_test)

    for i in range(1, epochs + 1):
        # Forward propagation for training set
        z_train = forward_prop(Xtrain, w, b)
        cost_train = cost_function(z_train, ytrain, w, lambd)

        # Backward propagation with regularization
        dw, db = back_prop(Xtrain, ytrain, z_train, w, lambd)

        # Gradient descent update
        w, b = gradient_descent_update(w, b, dw, db, learning_rate)

        if i % 10 == 0:
            costs_train.append(cost_train)

        # **Mean Squared Error (MSE) for training and testing**
        MSE_train = (1 / m_train) * np.sum(np.square(z_train - ytrain))

        # Testing on the test set
        z_test = forward_prop(Xtest, w, b)
        cost_test = cost_function(z_test, ytest, w, lambd)
        MSE_test = (1 / m_test) * np.sum(np.square(z_test - ytest))

        # Print progress every 10 epochs
        if i % 10 == 0:
            print(f'Epoch {i}/{epochs}:')
            print(f'Training cost: {cost_train} | Test cost: {cost_test}')
            print(f'MSE train: {MSE_train} | MSE test: {MSE_test}')

    # Plot the training cost over time
    plt.plot(costs_train)
    plt.xlabel('Iterations (per tens)')
    plt.ylabel('Training Cost')
    plt.title(f'Learning rate: {learning_rate}, Regularization lambda: {lambd}')
    plt.show()
```

```python
import numpy as np

class LogisticRegression:
    def __init__(self, learning_rate, n_iters, lambda_reg):
        self.lr = learning_rate
        self.n_iters = n_iters
        self.lambda_reg = lambda_reg  # Regularization strength
        self.weights = None
        self.bias = None
        self.losses = []
         
    # Sigmoid method
    def _sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def compute_loss(self, y_true, y_pred):
        # Binary cross-entropy loss with L2 regularization
        epsilon = 1e-9  # to avoid log(0)
        y1 = y_true * np.log(y_pred + epsilon)
        y2 = (1 - y_true) * np.log(1 - y_pred + epsilon)
        
        # Compute the L2 regularization term
        l2_loss = (self.lambda_reg / 2) * np.sum(np.square(self.weights))
        
        # Return the total loss: original loss + regularization
        return -np.mean(y1 + y2) + l2_loss

    def feed_forward(self, X):
        # Compute linear transformation (z = X * weights + bias)
        z = np.dot(X, self.weights) + self.bias
        A = self._sigmoid(z)  # Apply sigmoid to the output
        return A

    def fit(self, X, y):
        n_samples, n_features = X.shape

        # Initialize parameters
        self.weights = np.zeros(n_features)
        self.bias = 0

        # Gradient descent loop
        for _ in range(self.n_iters):
            # Compute the model's predictions
            A = self.feed_forward(X)

            # Compute loss and store it
            self.losses.append(self.compute_loss(y, A))  
            
            # Compute gradients for gradient descent
            dz = A - y  # Derivative of the loss function w.r.t. predictions (A)
            dw = (1 / n_samples) * np.dot(X.T, dz)  # Compute gradient for weights
            db = (1 / n_samples) * np.sum(dz)  # Compute gradient for bias
            
            # Add L2 regularization to the weights gradient
            dw += self.lambda_reg * self.weights  # Regularized gradient
            
            # Update parameters
            self.weights -= self.lr * dw
            self.bias -= self.lr * db
            
    def predict(self, X):
        # Predict the class labels
        threshold = 0.5
        y_hat = np.dot(X, self.weights) + self.bias
        y_predicted = self._sigmoid(y_hat)
        y_predicted_cls = [1 if i > threshold else 0 for i in y_predicted]
        
        return np.array(y_predicted_cls)

    def compute_mse(self, y_true, y_pred):
        # Compute Mean Squared Error (MSE)
        return np.mean((y_true - y_pred) ** 2)


regressor = LogisticRegression(learning_rate=0.0001, n_iters=1000, lambda_reg=0.01)
regressor.fit(Xtrain_set, Ytrain_set)


train_predictions = regressor.predict(Xtrain_set)
test_predictions = regressor.predict(Xtest_set)


train_mse = regressor.compute_mse(Ytrain_set, train_predictions)
test_mse = regressor.compute_mse(Ytest_set, test_predictions)

print(f"Training MSE: {train_mse}")
print(f"Testing MSE: {test_mse}")
```
